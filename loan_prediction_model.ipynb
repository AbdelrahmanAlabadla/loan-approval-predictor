```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings(action='ignore', category=UserWarning)
```

```python
df = pd.read_csv("loan_data[1].csv")


print(df.columns)
print(df.dtypes)
print(df.describe())
print(df.head())
print(df.isnull().sum())
print(df.shape)

df.columns=df.columns.str.strip()



le=LabelEncoder()
df["person_gender"]=le.fit_transform(df["person_gender"])
df["person_education"]=le.fit_transform(df["person_education"])
df = pd.get_dummies(df, columns=["person_home_ownership","loan_intent"], drop_first=True)


df=df.drop(["previous_loan_defaults_on_file","person_emp_exp","loan_int_rate",'loan_percent_income',
 "cb_person_cred_hist_length"],axis=1)

print("----------------------------------------")
print(df.columns.tolist())
```

```python
y=df["loan_status"]
df=df.drop(["loan_status"],axis=1)
x=df


scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)

x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)

model=LogisticRegression(class_weight="balanced",random_state=42,max_iter=1000)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)

f1score=f1_score(y_test,y_pred)
confusionmatrix=confusion_matrix(y_test,y_pred)
accuracy=accuracy_score(y_test,y_pred)




print("f1 score: ",f1score)
print("confuse matrix: ",confusionmatrix)
print("accuracy: ",accuracy)
```
